```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('data/Orange Quality Data.csv')
```

```python
fig, ax = plt.subplots()
size, harvest_time = df['Size (cm)'], df['HarvestTime (days)']

ax.grid(alpha=0.3)
ax.scatter(size, harvest_time)

ax.set_title('Relationship between Size and Harvest Time of Oranges', fontweight='bold')
ax.set_xlabel('Size (cm)', fontweight='bold')
ax.set_ylabel('Harvest Time (Days)', fontweight='bold')

mean_size = size.mean()
mean_harvest_time = harvest_time.mean()
correlation_coefficient = df['Size (cm)'].corr(df['HarvestTime (days)'])

summary_text = f"Mean Size: {mean_size:.2f}\nMean Harvest Time: {mean_harvest_time:.2f}\nCorrelation Coefficient: {correlation_coefficient:.2f}"
ax.text(0.95, 0.05, summary_text, transform=ax.transAxes, ha='right', va='bottom', bbox=dict(boxstyle='round,pad=0.5', fc='white', alpha=0.5))


plt.show()

```

```python
fig, ax = plt.subplots()
weight, size = df['Weight (g)'], df['Size (cm)']

ax.scatter(weight, size)

ax.set_title('Relationship between Weight and Size of Oranges', fontweight='bold')
ax.set_xlabel('Weight (g)', fontweight='bold')
ax.set_ylabel('Size (cm)', fontweight='bold')

plt.show()
```

```python
fig, ax = plt.subplots()
brix, quality = df['Brix (Sweetness)'], df['Quality (1-5)']

ax.scatter(brix, quality)

ax.set_title('Relationship between Brix and Quality', fontweight='bold')
ax.set_xlabel('Brix (Sweetness)', fontweight='bold')
ax.set_ylabel('Quality (1-5)', fontweight='bold')

summary_text = f'Mean Brix: {brix.mean():.2f}\nMean Quality: {quality.mean():.2f}\nCorrelation Coefficient: {brix.corr(quality):.2f}'
ax.text(0.95, 0.05, summary_text, transform=ax.transAxes, ha='right', va='bottom', bbox=dict(boxstyle='round,pad=0.5', fc='white', alpha=0.5))

plt.show()
```

```python
fig, ax = plt.subplots(figsize=(10,10))

variety = df.groupby('Variety')['Variety'].count().sort_values(ascending=False)

hbars = ax.barh(y=variety.index, width=variety, height=0.9, zorder=2)

ax.set_title('Count of Each Orange Variety', fontweight='bold')
ax.set_xlabel('Count', fontweight='bold')
ax.set_ylabel('Variety', fontweight='bold')

ax.bar_label(container=hbars, fmt='%.0f', label_type='center', color='w', fontweight='bold')
ax.set_xlim(right=25)
ax.grid(axis='x', ls='--', alpha=0.5, zorder=1)

plt.show()
```

```python
fig, ax = plt.subplots(figsize=(10,10))

orange_colors = df.groupby('Color')['Color'].count()

hex_colors = ['#FF5722',  # Deep Orange
              '#FFCCBC',  # Light Orange
              '#FF9800',  # Orange
              '#FF4500',  # Orange-Red
              '#FFD700']  # Yellow-Orange

ax.pie(x=orange_colors,
       explode=[0.05,0,0,0,0] ,
       labels=orange_colors.index, 
       autopct='%.2f%%', 
       textprops={'fontweight': 'bold'}, 
       wedgeprops={'linewidth': 1, 'edgecolor': 'black'},
       pctdistance=0.65,
       colors=hex_colors
       )
ax.set_title('Distribution of Orange Colors', fontweight='bold')
ax.legend(title='Colors', title_fontproperties={'weight': 'bold'}, loc='upper left')

```

```python
df['Blemishes'] = df['Blemishes (Y/N)'].str.get(0)
blemishes = df.groupby('Blemishes')['Blemishes'].count()
hex_colors = ['#FF5722',  # Deep Orange
              '#FFCCBC']

fig, ax = plt.subplots()

for i in range(len(blemishes)):
    bar_container = ax.bar(x=blemishes.index[i], height=blemishes[i], edgecolor='black', color=hex_colors[i], label=blemishes.index[i])
    ax.bar_label(container=bar_container, fmt="%d", label_type='center', fontweight='bold')

ax.set_title('Distribution of Blemishes', fontweight='bold')
ax.set_xlabel('Blemishes', fontweight='bold')
ax.set_ylabel('Count', fontweight='bold')
ax.legend(loc='upper right', title='Legend')

plt.show()
```
